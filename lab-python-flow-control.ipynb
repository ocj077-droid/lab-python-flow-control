```python
# Step 1: Define the list of products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
# Step 2: Create an empty inventory dictionary
inventory = {}
```

```python
# Step 3: Ask the user to input the quantity of each product
for product in products:
    quantity = int(input(f"Enter the quantity available for {product}: "))
    inventory['t-shirt'] = 10
```

```python
# Step 4: Create an empty set for customer orders
customer_orders = set()
```

```python
# Step 5: Ask the user to enter products until they choose to stop
while True:
    product_order = input("Enter the name of a product to order: ").lower()

    if product_order in products:
        customer_orders.add(product_order)
    else:
        print("This product is not available in the store.")

    another = input("Do you want to add another product? (yes/no): ").lower()
    if another != "yes":
        break
```

```python
# Step 6: Print the products in the customer_orders set
print("\nProducts ordered:")
for product in customer_orders:
    print(product)
```

```python
# Step 7: Calculate order statistics
total_products_ordered = len(customer_orders)
percentage_ordered = (total_products_ordered / len(products)) * 100

order_status = (total_products_ordered, percentage_ordered)
```

```python
# Step 8: Print the order statistics
print("\nOrder Statistics:")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]}%")
```

```python
# Step 9: Update the inventory only for ordered products
for product in customer_orders:
    if inventory[product] > 0:
        inventory[product] -= 1
```
